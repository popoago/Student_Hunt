## Missing Value treatment

```python
train.apply(lambda x: sum(x.isnull()))
```

```python
mean = train.groupby("Date").mean()
```

```python
for i in train.columns:
    def impute(para):
        val,date=para
        if val!=val:
            return mean[i][date]
        else:
            return val 
    train[i] = train[[i,"Date"]].apply(impute,axis=1)
```
